**REINFORCEMENT LEARNING**


It is a type of machine learning method where the AI agent is trained to learn with time.

AI agent is developed in such a way that it iteratively interacts with the environment and perform some basic action s.

For every action, feedback is given to the AI agent.

The feedback is in such a form, that the agent gets a positive reward for every positive action and it gets penalty as s feedback for every negative action.

The AI agent aim to perform those actions that lead to getting the maximum of the rewards(positive feedback) as an out[put.

In reinforcement learning(RL), the environment is a critical component that interacts with the agent. It representd everything outside the agent that the agent can percieve and act upon. The environment provides the agent with feedback in the form of states and rewards, whuch the agent uses to learn and improve its policy.

1-Learning from interaction with an environment to acheive some long term goal that is related to the state of the environment

2-The goal is defined by reward signal, which must be maximized

3-Agent must be able to partially/fully sense the environment state and take

actions to influence the environment state

4-Learning depending on the changes occuring in the environment

5-it is employed by various softwares and machines to find the best possible behaviour or path it should take in a specific situation

Goal: To acheive the best learning/Result

**Main points in Reinforcement Learning**

Input: The input should be an initial state from which the model will start

Output: There are many possible outputs as they are a variety of solutions to a particular problem

Training: The training is based upon the input, The model will return a state and the user will decide to reward or punish the model based on its output

The model keeps continues to learn

The best solution is decided based on the maximum reward.

Types of Reinforcement learning:
